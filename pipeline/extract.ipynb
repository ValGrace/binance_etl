```python
import pandas as pd
import requests as req
```

```python
# URLS
price_url = 'https://api.binance.com/api/v3/ticker/price'
trades = 'https://api.binance.com/api/v3/trades?symbol=ETHBTC&limit=5'
```

```python
def get_prices():
    res = req.get(price_url)
    res.raise_for_status()

    data = res.json()
    print(data)
    return data
```

```python
prices = get_prices()
```

```python
prices
```

```python
def get_trades():
    resp = req.get(trades)
    resp.raise_for_status()

    data = resp.json()
    print(data)
```

```python
get_trades()
```

```python
def order_book():
    res = req.get('https://api.binance.com/api/v3/depth?symbol=BTCUSDT&limit=5')
    res.raise_for_status()

    data = res.json()
    print(data)
```
